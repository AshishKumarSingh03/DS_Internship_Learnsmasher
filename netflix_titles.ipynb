```python
import matplotlib.pyplot as plt

data['country'].value_counts().head().plot(kind='bar', color=['pink','green', 'red', 'purple','blue'])
plt.title("Top 5 country")
plt.xlabel("country")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt

data['release_year'].value_counts().head().plot(kind='bar', color=['red', 'blue', 'green', 'orange', 'purple'])
plt.title("Top 5 Release Years")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt

data['director'].value_counts().head().plot(kind='bar', color=['pink','green', 'red', 'purple','blue'])
plt.title("Top 5 director")
plt.xlabel("director")
plt.ylabel("Count")
plt.show()
```

```python
plot = plt.pie(data.type.value_counts(),explode=(0.025,0.025), labels=data.type.value_counts().index, colors=['yellow','green'],autopct='%1.1f%%', startangle=180)
plt.show()
```

```python
filtered_countries = data.set_index('title').country.str.split(', ', expand=True).stack().reset_index(level=1, drop=True);
filtered_countries = filtered_countries[filtered_countries != 'Country Unavailable']
plt.figure(figsize=(13,7))
plot = sns.countplot(y = filtered_countries, order=filtered_countries.value_counts().index[:20])
plt.title('Top 20 Countries Contributor on Netflix')
plt.xlabel('Titles')
plt.ylabel('Country')
plt.show()
```

```python
filtered_directors = data[data.director != 'No Director'].set_index('title').director.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
plt.figure(figsize=(13,7))
plt.title('Top 20 Director Based on  number of Titles')
sns.countplot(y = filtered_directors, order=filtered_directors.value_counts().index[:20], palette='Greens')
plt.show()
```

```python
filtered_genres = data.set_index('title').listed_in.str.split(', ', expand=True).stack().reset_index(level=1, drop=True);
plt.figure(figsize=(10,10))
g = sns.countplot(y = filtered_genres, order=filtered_genres.value_counts().index[:20])
plt.title('Top 20 Genres on Netflix')
plt.xlabel('Titles')
plt.ylabel('Genres')
plt.show()    
```
